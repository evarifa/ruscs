```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def draw_hexagon(ax, center, size, edge_color='gold', face_color='orange'):
    angle = np.linspace(0, 2*np.pi, 7)
    x = center[0] + size * np.cos(angle)
    y = center[1] + size * np.sin(angle)
    ax.fill(x, y, face_color, edgecolor=edge_color, linewidth=3.5) 

def draw_honeycomb(rows, cols, size=1, probability=0.8, spacing=0.2):
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.set_aspect('equal')

    for row in range(rows):
        for col in range(cols):
            x = col * size * np.sqrt(3) * 1
            y = row * size * np.sqrt(3) * (1 + spacing) + size
            if col % 2 == 1:
                y += size * np.sqrt(3) * (1 + spacing) / 2
            if np.random.rand() < probability:
                draw_hexagon(ax, (x, y), size)

    ax.set_xlim(-size, cols * size * np.sqrt(3) * 1.25)
    ax.set_ylim(-size, rows * size * np.sqrt(3) * (1 + spacing) + size)
    ax.axis('off')
    plt.show()
    
draw_honeycomb(5, 7, probability=0.8, spacing=0.2)

```
